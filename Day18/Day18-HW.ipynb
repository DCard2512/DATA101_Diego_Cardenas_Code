```python
model = DF[['line_breaks','spam']]
model['line_breaks'].value_counts()
show(model.groupby('spam')['line_breaks'].mean())
```

```python
X = model['line_breaks'].values.reshape(-1,1)
y = model['spam']

LM = LogisticRegression()
LM.fit(X,y)
```

```python
model['prediction'] = LM.predict(X)

fig = px.scatter(model,x='line_breaks',y='spam')

fig.show()
```
